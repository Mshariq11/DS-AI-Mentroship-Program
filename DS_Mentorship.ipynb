```python
# Assignment how to combine 3 columns in one column? make a separate data of everyting based on sex (Male, Female, and Transgender)

# First Convert the dtypes from integer columns to strings:
pakpop['MALE (RURAL)'] = pakpop['MALE (RURAL)'].astype(str)
pakpop['FEMALE (RURAL)'] = pakpop['FEMALE (RURAL)'].astype(str)
pakpop['TRANSGENDER (RURAL)'] = pakpop['TRANSGENDER (RURAL)'].astype(str)

# Then, Create a new column by combining three columns:
pakpop['combined_column'] = pakpop['MALE (RURAL)'].str.cat([pakpop['FEMALE (RURAL)'], pakpop['TRANSGENDER (RURAL)']], sep=' ')

# Separate data based on sex
male_data = pakpop[pakpop['combined_column'].str.contains('MALE')]
female_data = pakpop[pakpop['combined_column'].str.contains('FEMALE')]
transgender_data = pakpop[pakpop['combined_column'].str.contains('TRANSGENDER')]
```
